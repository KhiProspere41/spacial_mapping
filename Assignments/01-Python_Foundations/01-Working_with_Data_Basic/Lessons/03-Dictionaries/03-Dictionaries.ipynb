# 📘 03 - Python Dictionaries

Dictionaries store key-value pairs for fast lookups.

## ✅ Learning Goals
- Create and access dictionary items
- Use dictionary methods like `.get()` and `.items()`

```python
# 📚 Create a Dictionary
student = {'name': 'Alex', 'age': 21, 'major': 'CS'}
print(student)
```

```python
# 🔍 Access Values
print(student['name'])
print(student.get('gpa', 'N/A'))
```

```python
# 🔁 Loop Over Dictionary
for key, value in student.items():
    print(key, '->', value)
```

## ✏️ Your Turn
Create a dictionary to store information about your favorite book.

```python
#Create a Dictionary
book = {'title': 'Diary of a Wimpy Kid: Rodrick Rules', 'author': 'Jeff Kinney', 'year': 2008}
print(book)

#Access Values
print(book['title'])
print(book.get('pages', 'N/A')) # 'pages' doesn't exist, so it will print 'N/A'

#Loop Over Dictionary
for key, value in book.items():
    print(key, '->', value)
```
