# 📘 02 - Control Flow

Learn how to use conditional logic in Python.

## ✅ Learning Goals
- Write `if`, `elif`, `else` statements
- Understand truthy/falsy values
- Combine conditions using `and`, `or`, `not`

```python
# 🔧 Basic example
age = 20
if age >= 18:
    print('You are an adult.')
else:
    print('You are a minor.')
```

## 🔀 Multiple Branches

```python
temp = 72
if temp > 90:
    print('Too hot!')
elif temp < 60:
    print('Too cold!')
else:
    print('Just right.')
```

Use `input()` to ask a user their grade and print a message based on value.

```python
# Ask the user for their grade
grade = int(input("Enter your grade (0-100): "))

#Check the grade and print a message
if grade >= 90:
    print("You got an A!")
elif grade >= 80:
    print("You got a B!")
elif grade >= 70:
    print("You got a C!")
elif grade >= 60:
    print("You got a D!")
else:
    print("You got an F.")
```
